```python
#features with nan values
features_with_nan=[features for features in df.columns if df[features].isnull().sum()>0]
features_with_nan
```

```python
from percent import percentage
percentage(df)
```

```python
## lets plot some amezing diagrams 

for i in features_with_nan:
    data=df.copy()
    #make a variable that indicates 1 if the observation is missing and 0
    data[i]=np.where(data[i].isnull(),1,0)
    
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.title(i)
    plt.show()
```

```python
#Tamporable variables(Eg.Date time variables)
year_feature=[i for i in numerical_variables if 'Yr' in i or 'Year' in i]
year_feature
```

```python
for i in year_feature:
    print(i,df[i].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YrSold')
plt.ylabel('Median House Price')
plt.title('House Price vs YrSold',)
```

```python
### Here we will comppare diff all year feature with salse prise
data=df.copy(0)
for i in year_feature:
    if i!='YrSold':
        data[i]=data['YrSold']-data[i]
        
        plt.scatter(data[i],data['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('SalePrice')
        plt.show()
    
```

```python
## Numerical vriables
##1 continues variable and Discretw Vriable

discrete_feature=[i for i in numerical_variables if len(df[i].unique())<=25]
print(len(discrete_feature))
```

```python
df[discrete_feature].head()
```

```python
## lets find the relationship between Diiscrete and sales price
data=df.copy()
for i in discrete_feature:
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# there is a relatoion between disctrete features and salesprice
```

```python
## continues variable 
continues_feature=[i for i in numerical_variables if i not in discrete_feature+year_feature]
print(len(continues_feature))
```

```python
## lets analyse the continues values by creating histograms to understand the distribution
data=df.copy()
for i in continues_feature:
    data[i].hist(bins=25)
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.title(i)
    plt.show()
```

# EDA part2

```python
#we will be using logarithemic transformation
data=df.copy()
for i in continues_feature:
    if 0 in df[i].unique():
        pass
    else:
        df[i]=np.log(data[i])
        plt.scatter(data[i],data['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('SalePrice')
        plt.title(i)
        plt.show()
```

```python
#outliers
for i in continues_feature:
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data.boxplot(column=i)
        plt.ylabel(i)
        plt.title(i)
        plt.show()
```
